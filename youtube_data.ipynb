```python
m = widgets.Dropdown(
    options=channel['Genre'].unique(),
    value='Sports',
    description='Number:',
    disabled=False,
)
def graph(a):
  sports = channel[channel['Genre'] ==a]
  newsports =  sports.groupby(by = 'channelTitle').sum().sort_values(by = 'view_count',ascending = False).head(10)
  sns.barplot(x = newsports.index,y = newsports['view_count'])
  plt.title(a)
  plt.xticks(rotation=90)
  plt.show()
  
```

```python
widgets.interact(graph,a = m)
```
